```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
combined_df = pd.read_csv('all_regions_full_data.csv')
display(combined_df)
```

## Create a dataframe using the ` .groupby()` code to focus on the crime type and totals per 'Postcode area'

```python
postcode_crimes_total = combined_df.groupby(['Postcode area', 'Crime type', 'Price', 'Total Population']).agg({'Latitude': 'mean', 'Longitude': 'mean', 'Crime type': 'count'})
postcode_crimes_total.columns = ['Latitude', 'Longitude', 'Total Count']
postcode_crimes_total.reset_index(inplace=True)
```

```python
postcode_crimes_total_df = postcode_crimes_total

display(postcode_crimes_total_df)
```

## Create a dictionary to group the 'Crime Type' into 'Severity' groupings
## This is then mapped using ` for` loops and ` groupby` onto a separate dataframe

```python
# Define the mapping of crime types to severity levels
severity_mapping = {
    'High Severity': ['Violence and sexual offences','Robbery','Criminal damage and arson'],
    'Medium Severity': ['Burglary', 'Vehicle crime',  'Possession of weapons'],
    'Low Severity': ['Public order', 'Other theft', 'Drugs', 'Bicycle theft', 'Shoplifting', 'Theft from the person', 'Other crime']
}

# Reverse mapping to get severity level for each crime type
reverse_mapping = {crime: level for level, crimes in severity_mapping.items() for crime in crimes}

# Group the crime types by severity
postcode_crimes_total_df['Severity'] = postcode_crimes_total_df['Crime type'].map(reverse_mapping)

# Group by 'Postcode area' and 'Severity', then calculate total counts
postcode_crimes_total_by_severity = postcode_crimes_total_df.groupby(['Postcode area', 'Severity', 'Price', 'Total Population']).agg({'Latitude': 'mean', 'Longitude': 'mean', 'Total Count': 'sum'}).reset_index()

display(postcode_crimes_total_by_severity)

postcode_crimes_total_by_severity.to_csv('Crime_severity.csv')
```

```python
# Group by 'Postcode area' and aggregate the 'Total Count' column by sum
total_count_df = postcode_crimes_total_by_severity.groupby(['Postcode area', 'Price', 'Total Population', 'Latitude', 'Longitude']).agg({'Total Count': 'sum'}).reset_index()

# Define the display format for float values
pd.options.display.float_format = '{:.2f}'.format

# Display the total count dataframe
display(total_count_df)

# Save the total count dataframe to a CSV file
total_count_df.to_csv('total_crime_severity.csv', index=False)
```
