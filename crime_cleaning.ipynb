# Concatting streets databases


```python

import pandas as pd 
import os

folder_path = r'All_directory'
#out_file is the exported csv file 
output_file = 'merged_street_data.csv'
 
def merge_street_data(folder_path, output_file):
    #Creates an empty list to store the csv files
    main_csv = []
 
    #Traverse the directory tree rooted at folder_path
    for root,dirs,files in os.walk(folder_path):
        #Iterate over the files in current directory
        for file in files:
            #Checks if the file ends with 'street.csv'
            if file.endswith('street.csv'):
                #Construct the full path to the csv file
                file_path = os.path.join(root,file)
                #read the csv file into a Dataframe
                csv = pd.read_csv(file_path)
                #Append the Dataframe into the empty list
                main_csv.append(csv)
    #Merges all Dataframes in the list into a single Data frame
    #ignore_index makes sure that the merged_csv has a reset index
    merged_csv = pd.concat(main_csv, ignore_index = True)
    #Saves the merged_csv
    #index = False so that it does not have an index in the first column
    merged_csv.to_csv(output_file,index = False)
    #Print a message to say if it was saved successfully
    print('Saved successfully!')
merge_street_data(folder_path, output_file)
```

# House Prices


```python
house_prices = pd.read_csv('house_prices.csv')
display(house_prices)
```

## Filtering out the 'Postcode' columns to get the 'Postcode area'

```python
# Split the 'Postcode' column and take the first part
house_prices['Postcode district'] = house_prices['Postcode'].str.split().str[0]

# Display the DataFrame with the new column
display(house_prices)
house_prices.to_csv('house_prices.csv')

```

```python
import numpy as np

# Define a function to extract the alphabetic characters from the postcode district
def extract_postcode_area(postcode):
    if isinstance(postcode, str):  # Check if the value is a string
        return ''.join(filter(str.isalpha, postcode))
    else:
        return np.nan  # Return NaN if the value is not a string

# Apply the function to create the new 'Postcode area' column
house_prices['Postcode area'] = house_prices['Postcode district'].apply(extract_postcode_area)

# Display the first few rows of the DataFrame to verify the new column
display(house_prices)

house_prices.to_csv('house_prices_filtered.csv')
```

```python
filtered_house = pd.read_csv('house_prices_filtered.csv')

# Define the list of postcode areas
postcode_areas = ['HA', 'MK', 'LU', 'SG', 'HP', 'AL', 'SE', 'IG', 'E', 'SW',
                  'EN', 'KT', 'WD', 'TW', 'UB', 'NW', 'RM', 'N', 'OX', 'RG', 'CR',
                  'CM', 'SS', 'CB', 'CO', 'EC', 'GU', 'RH', 'SM', 'SL', 'TN', 'DA',
                  'BR', 'WC', 'W', 'NN', 'SN', 'GL']

# Filter the DataFrame to include only the data from the specified postcode areas
filtered_house_postcode = filtered_house[filtered_house['Postcode area'].isin(postcode_areas)]

# Display the filtered DataFrame
display(filtered_house_postcode)

filtered_house_postcode.to_csv('house_prices_filtered.csv')
```

## Create a separate dataframe that consists of the average price of the 'Postcode area' 


```python
# Group the data by 'Postcode area' and calculate the mean price for each area
mean_prices = filtered_house_postcode.groupby('Postcode area')['Price'].mean().reset_index()

# Set the display format for floating-point numbers
pd.options.display.float_format = '{:.0f}'.format

# Display the mean prices DataFrame with the adjusted format
display(mean_prices)

mean_prices.to_csv('Average House Prices.csv')
```

```python
mean_prices = pd.read_csv('Average House Prices.csv')
filled_data = pd.read_csv('merged_postcodes.csv')

display(filled_data.columns)
display(mean_prices.columns)

# Merge the filled_data with mean_prices on the 'Postcode area' column
merged_data = pd.merge(filled_data, mean_prices, on='Postcode area', how='left')

# Display the first few rows of the merged DataFrame
display(merged_data)

merged_data.to_csv('all_regions_full_data.csv')
```

## Create a separate dataframe for 'Total Population' for each 'Postcode area' 

```python
postcode_areas = ['HA', 'MK', 'LU', 'SG', 'HP', 'AL', 'SE', 'IG', 'E', 'SW',
                  'EN', 'KT', 'WD', 'TW', 'UB', 'NW', 'RM', 'N', 'OX', 'RG', 'CR',
                  'CM', 'SS', 'CB', 'CO', 'EC', 'GU', 'RH', 'SM', 'SL', 'TN', 'DA',
                  'BR', 'WC', 'W', 'NN', 'SN', 'GL']

# Initialize an empty list to store the data
data = []
# Iterate over each postcode area
for area in postcode_areas:
    # Calculate the total population for the current postcode area
    total_population = merged_data[merged_data['Postcode area'] == area]['Population'].sum()
    # Append the postcode area and total population to the data list as a dictionary
    data.append({'Postcode area': area, 'Total Population': total_population})

# Create a DataFrame from the collected data
total_population_df = pd.DataFrame(data)
# Display the DataFrame
display(total_population_df)

total_population_df.to_csv('total_population.csv')
```

```python
population_data = pd.read_csv('total_population.csv')
display(population_data)

filled_data = pd.read_csv('all_regions_full_data.csv')
display(filled_data.columns)
```

# Final merged dataframe


```python
# Merge the dataframes with explicit suffixes
merged_data = pd.merge(filled_data, population_data, on='Postcode area', how='inner', suffixes=('_filled', '_population'))

# Drop the duplicate columns
merged_data = merged_data.drop(['Unnamed: 0_filled', 'Unnamed: 0_population'], axis=1)

# Display the modified dataframe
display(merged_data)


merged_data.to_csv('all_regions_full_data.csv')
```
